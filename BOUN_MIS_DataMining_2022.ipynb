### Discretization

#### Discretization: qcut function

```python
# qcut: Quantile-based discretization function
# all bins (both low and high) will have (if the whole is even, or else roughly) the same number of observations
data['tsbin'] = pd.qcut(data['TotalSales'], q=2, labels=["low", "high"])
# param q is set to 2, which means, TotalSales data values is devided into 2 equal bins (categories)

data.columns
data.head()

print(data['tsbin'].value_counts())
print(data['tsbin'].describe()) 

# So now, we can find the edges of low and high in tsbin
# Let's figure it out by finding the max and min respectively
max_of_tsbin_low = data[data.tsbin=='low'].TotalSales.max()
print(f"The max. of 'low' category in tsbin: {max_of_tsbin_low}")

min_of_tsbin_high = data[data.tsbin=='high'].TotalSales.min()
print(f"The min. of 'high' category in tsbin: {min_of_tsbin_high}")

# qcut function by giving quantile intervals
# we covered all the data from %0 (first q element) to %100 (last q element)
# quantile intervals: (%0,%20], (%20,%60], (%60,%100]
data['tsbin2'] = pd.qcut(data['TotalSales'], q=[0,0.2,0.6,1.0], labels=["low", "normal", "high"])

print("Information for: tsbin2:")
print(data['tsbin2'].value_counts())
print(data['tsbin2'].describe())
```

#### Discretization: cut function

```python
# cut: it is used to specifically define the bin edges
# the intervals: (0,10],(10,30],(30,250] -> 3 different bins
#                 low    normal   high
# 0 is NOT included in category: "low"
# 0.1 is included in category: "low"
data['tsbin3'] = pd.cut(data['TotalSales'], bins=[0,10,30,250], labels=["low", "normal", "high"])

print(data['tsbin3'].value_counts())
print(data['tsbin3'].describe())

data.head()
```

### PCA

#### Preprocessing: Prepare the features

```python
print(data.columns)

# Creating the feature dataFrame
feature = data[['TotalDrinkSales', 'TotalFoodSales', 'TotalNonConSales']]

print(feature.columns)
```

#### Preprocessing: Standard Scaling (z-normalization)

```python
# StandardScaler is used for normalization 
from sklearn.preprocessing import StandardScaler

# Insight: TotalFoodSales has a dominance effect unless we normalize the data

# Initiate StandardScaler object and transform the features into scaled values in an array form
X = StandardScaler().fit_transform(feature)

print(f"Mean: {X.mean()}")
print(f"Std: {X.std()}")

print(X)
print(type(X))

# Convert X from array type to Dataframe type for ease of use
X_dataframe = pd.DataFrame(X, columns=feature.columns)

# See that the correlation does not change when normalization applied
print(feature.corr())
print(X_dataframe.corr())
```

#### Compute PCA

```python
from sklearn.decomposition import PCA

# Initiate the model, PCA object
pca = PCA(n_components=3)

# Apply dimensionality reduction on X
z = pca.fit_transform(X)

# Reminder: data (original version, all columns) -> feature (only chosen columns) -> X (scaled version of feature)

# z is the result of PCA, namely transformed values, i.e. coordinates of the individuals (observations)
print(z)
# shape of z will be same as the shape of X (same number of rows and columns)
print(z.shape)

variance = pca.explained_variance_ratio_
print(f"Proportion of Variance: {variance}")
# PC1 is able to explain the variance of the original data in ~%90 similarity

# the below code result is the same as above
# print(f"Proportion of Variance: {pca.explained_variance_ratio_}")

# Each row corresponds to a principal component (PC1, PC2, PC3)
# Each column corresponds to an attribute (TotalDrinkSales, TotalFoodSales, TotalNonConSales)
print(pca.components_)
# We can see in the first row that, PC1 contains %56,%58,%57 from all the original attributes (variables) respectively
# So PC1 can be used to represent all the original attributes

# In above, proportions of each feature are called "Loading Scores"
# Eigen value for PC1 = SS(distances for PC1)
# Singular value for PC1 = sqrt(Eigen value for PC1)

# pca.components_ gives the Eigen Vectors in each row for each principal component
# sum of squares of those gives 1, because eigen vector is a unit vector
print(f'Unit: {pow(pca.components_[0,0],2) + pow(pca.components_[0,1],2)  + pow(pca.components_[0,2],2)}' )

# That gives the Eigen Values
# And the total of the variance is 3: [2.70108964 0.21336781 0.08583438]
print(f'Eigen Values: {pca.explained_variance_}')

# X_dataframe
z_dataframe = pd.DataFrame(z, columns=['PC1','PC2','PC3'])

# concat function is used to concatenate two dataframes
corr_data = pd.concat([X_dataframe, z_dataframe], axis=1)

# We see that principal components are not correlated each other
corr_data.corr()
```

```python
import numpy as np

# this is the Loading Scores (PC Loadings) or Eigen Vectors
V = pca.components_
# compute the transpose of PC Loadings
Vt = V.transpose()

result = np.dot(X, Vt)
# z = XVt / XVt is equal to z (PCA Scores)
print(result)
print(z)
```

### Frequent Pattern Mining

#### Install & Import Libraries

```python
# Frequent Pattern Mining
# Install and import libraries
!pip install apyori

import pandas as pd
from apyori import apriori
```

#### Create and run the helper functions

```python
# Helper Utility Functions
# You can re-use those methods
# You are not responsible for knowing the details or creating those methods from scratch

# In order for the following computations to run without an error, this code cell/panel need to be run once - so press play button

# used for converting pandas Dataframe to transactional format (list of lists)
def convertToTransactions(data):
  rowCount = data.shape[0]
  columnCount = data.shape[1]

  transactions = []
  for i in range(0, rowCount):
    transaction = []
    for j in range(0, columnCount):
      value = str(data.loc[i, j])
      if value!='nan':
        transaction.append(value)
    transactions.append(transaction)
  return transactions

# used for converting the transactional format (list of lists) into a pandas Dataframe, by sorting
def convertToDataFrame(results, sortBy=None):
  rules = []
  for item in results:     
    for rule in item.ordered_statistics:
      items = item[0]
      support = item.support
      lhs = str([x for x in rule.items_base])  
      arrow =  " -> " 
      rhs = str([x for x in rule.items_add])  
      confidence = rule.confidence
      lift = rule.lift
      rules.append([items, support,lhs,arrow,rhs,confidence,lift])
  output = pd.DataFrame(rules, columns=['items', 'support', 'lhs', 'direction', 'rhs', 'confidence', 'lift'])
  
  if sortBy:
    output = output.sort_values(by=[sortBy], ascending=False) # another option than assigning: inplace=True
  return output

# used for only generating the frequent itemsets by sorting, if you do not want to see the rules
def getFrequentItemset(transactions, support, length, sortBy=None):
  if isinstance(transactions, pd.DataFrame):
    transactions = convertToTransactions(transactions)

  records = apriori(transactions, min_support = support, max_length = length)
  results = list(records)

  rules = []
  for item in results:     
    for rule in item.ordered_statistics:
      items = item[0]
      support = item.support
      rules.append([items,support])
  output = pd.DataFrame(rules, columns=['items', 'support'])
  
  if sortBy:
    output = output.sort_values(by=[sortBy], ascending=False) # another option than assigning: inplace=True
  return output
```

#### Preprocessing: Generate Transactions

```python
# In order to use the existing apriori function, we need to first convert the imported data to transactional format (list of lists)
# Let's use our helper utility function to achieve this
transactions = convertToTransactions(data)

print(transactions)
```

#### Apriori

```python
# Apply apriori with parameters
records = apriori(transactions, min_support=0.01, min_confidence=0.2, min_lift=0.8, max_length=5)

result = list(records)

# This is for adjusting the output width for readability purposes
pd.set_option('max_colwidth', None)

display(pd.DataFrame(result))
```
